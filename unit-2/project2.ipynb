### Set up data frame:

```python
import pandas as pd 
colMajDta = pd.read_csv("college-major-data.csv", index_col = 0)
df = pd.DataFrame(colMajDta)
df
```

### Observations from Unmanipulated Data

Find and remove all the data with NaN or empty values to get more accurate readings.

```python
df[df.isnull().any(axis=1)] #https://datatofish.com/rows-with-nan-pandas-dataframe/
df = df.dropna()
```

## Observations from Data by Category

```python
df_ctgrs = df.groupby('Major_category')
df_ctgrs 
```

```python
#How many categories are there?
df_ctgrs.ngroups 
```

```python
#What are the categories and how many majors are in each?
size = df_ctgrs.size()
```

```python
#create a table to show data by category 
data_by_category = pd.concat([size],axis=1)
data_by_category
```

```python
#How does unemployment vary between categories? / What is the mean unemployment rate of each category?
avg_unemploy = df_ctgrs.Unemployment_rate.mean()
avg_unemploy
```

```python
df_ctgrs.Unemployment_rate.agg(['mean','max']).sort_values('mean', ascending = False).head(5)
```

```python
df_ctgrs.Unemployment_rate.agg(['mean','max']).sort_values('mean', ascending = False).tail(5)
```

```python
#Add a column to new table for unemployment rate
data_by_category['avg_unemploy'] = avg_unemploy
data_by_category.sort_values('avg_unemploy', ascending = False)
```

The above data shows that major categories such as Arts, Law & Public Policy, and Social Science have the highest average unemployment rates around 9%, compared to Physical Sciences at a low of 5%

```python
#Which categories have more men than women?
#start by finding total people
num_people = df.groupby('Major_category').Total.sum()
```

```python
#find total num of men
num_men = df.groupby('Major_category').Men.sum()
```

```python
#find percentage
percent = (num_men/num_people)*100
percent
```

```python
#Add column to data frame
data_by_category['percent_men'] = percent
data_by_category.sort_values('percent_men', ascending = False)
```
