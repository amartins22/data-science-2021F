# House Prices

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
train_houses = pd.read_csv('train.csv')
test_houses = pd.read_csv('test.csv')
test_houses
```

```python
train_houses.info()
```

```python
test_houses.info()
```

```python
house_features = ['LotArea','BedroomAbvGr','TotRmsAbvGrd','OverallQual','OverallCond','1stFlrSF','2ndFlrSF','FullBath','GarageCars','Fireplaces']
#

y = train_houses.SalePrice
X = train_houses[house_features]

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
model = RandomForestRegressor(random_state=1)
model.fit(train_X, train_y)
```

```python
preds = model.predict(val_X)
mae= mean_absolute_error(val_y, preds)

print(mae)
```

### Try it on all train data

```python
full_model = RandomForestRegressor(random_state=1)
full_model.fit(X,y)

full_preds = full_model.predict(X)

print(preds)
```

### Make df to see results with training data

```python
output = pd.DataFrame({'Id': train_houses.Id,
                      'SalePrice': train_houses.SalePrice,
                      'Prediction': full_preds})
output
#output.to_csv('submission.csv', index=False)
```

## Try it on Test Data

```python
test_houses['GarageCars'] = test_houses['GarageCars'].fillna(0)
test_X = test_houses[house_features]

test_preds = full_model.predict(test_X)

test_output = pd.DataFrame({'Id': test_houses.Id,
                      'SalePrice': test_preds})
test_output
test_output.to_csv('submission.csv', index=False)
```
