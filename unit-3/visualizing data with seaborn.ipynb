# Visualizing Data with Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as robLowe
import pandas as pd
```

```python
people = pd.read_csv("people.csv", index_col = 0)
people
```

```python
robLowe.lmplot(x='Number of Friends', y='GPA',data=people)
plt.show()
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col = 0)
reviews
```

```python
reviews = reviews.loc[reviews.points == 100]
reviews
```

```python
num_reviews = reviews.value_counts('country')
num_reviews
```

```python
robLowe.barplot(x=num_reviews.index, y=num_reviews)

plt.grid(axis = 'y')
plt.xlabel('Country')
plt.ylabel('Number of Reviews')
plt.title('Number of Perfect Reviews per Country', fontsize = 18)

plt.show()
```

```python
robLowe.barplot(x=num_reviews, y=num_reviews.index)

plt.grid(axis = 'x',zorder=0)
plt.ylabel('Country')
plt.xlabel('Number of Reviews')
plt.title('Number of Perfect Reviews per Country', fontsize = 18)

plt.show()
```
