# Text Processing Project

```python
import string
book = "crime-punish"
```

### Read Text File

```python
with open (book +".txt", "r") as myfile:
    txt=myfile.read().replace('\n',' ')
```

### Cleaning Text

#### Making a new list of all words separated by a space and removing puntuation

```python
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

for ele in txt:
    if ele in punc:
        txt = txt.replace(ele, "")

#https://www.geeksforgeeks.org/python-remove-punctuation-from-string/
```

```python
txt = txt.split(' ')
txt_list = [i for i in txt if i != '']
#txt_list
```

### Finding Next Words for Each Word

Make list a set to find unique words

```python
txt_set = set(txt_list)
txt_set
```

Removing words that are not from the actual story

```python
# txt_set = txt_set.remove('CHAPTER')
# type(txt_set)
```

Making the set of unique words a dictionary

```python
txt_dict = dict.fromkeys(txt_set, '') #https://www.geeksforgeeks.org/python-convert-a-set-into-dictionary/
txt_dict
```

Making lists of all the following words for each word and adding them to the dictionary

```python
def get_next_words(word):
    next_words =[]
    for i in range(len(txt_list)-1):
       if txt_list[i] == word:
          next_words.append(txt_list[i+1])
    return next_words
```

```python
for word in txt_set:
    txt_dict[word] = get_next_words(word)
```

```python
txt_dict
```

Find the most recurring following word for each word

```python
def most_frequent_word(list):
    return max(set(list), key=list.count)

#https://stackoverflow.com/questions/1518522/find-the-most-common-element-in-a-list
```

Making a final dictionary with every word and all the most frequent following words

```python
final_dict = dict.fromkeys(txt_set,'')

for word in txt_set:
    final_dict[word] = most_frequent_word(get_next_words(word))

final_dict
```
