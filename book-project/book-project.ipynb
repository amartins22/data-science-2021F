### 2. Clean Text

#### Making a new list of all words separated by a space, removing puntuation, and converting everything to lowercase

```python
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

for ele in txt:
    if ele in punc:
        txt = txt.replace(ele, "")

#https://www.geeksforgeeks.org/python-remove-punctuation-from-string/
```
